```python
import numpy as np; # importando a biblioteca numpy como np
import matplotlib.pyplot as plt; # importando a biblioteca matplot.pyplot como plt

t_max = 1; # tempo
fs = 100; # frequência de amostragem
f_m = 10; # frequência da mensagem

# SNR = 0;
 
x = np.linspace(0, t_max, t_max * fs + 1) # criando a quandidade de pontos no tempo necessarios

def generateSignals(SNR):
    sinal = np.sin(2 * np.pi * f_m * x)
    S = sum(abs(sinal)**2)/len(sinal);
    S_db = 10*np.log10(S);
    N_db = S_db - SNR;
    N = 10**(N_db/10)
    mean = 0;
    std = np.sqrt(N);
    ruido = np.random.normal(mean, std, size=len(sinal));
    sinal_recebido = sinal+ruido;

    plt.subplot(3, 1, 1); # definindo nº de linhas, nº de colunas e index do grafico do sinal de entrada 
    plt.plot(x, sinal, 'green'); # definindo valores do eixo "x" = x e "y" = sinal_0dB e a cor da linha do grafico do sinal de entrada
    plt.xlabel("Tempo"); # definindo descrição para o eixo x
    plt.ylabel("Amplitude"); # definindo descrição para o eixo y
    plt.title("Sinal de entrada para SNR=" + str(SNR) + "dB"); # definindo título para o gráfico do sinal de entrada

    plt.subplot(3, 1, 2); # definindo nº de linhas, nº de colunas e index do grafico do sinal do ruído
    plt.plot(x, ruido, 'red'); # definindo valores do eixo "x" = x e "y" = ruido_0db e a cor da linha do grafico do sinal do ruído
    plt.xlabel("Tempo"); # definindo descrição para o eixo x
    plt.ylabel("Amplitude"); # definindo descrição para o eixo y
    plt.title("Ruído para SNR=" + str(SNR) + "dB"); # definindo título para o gráfico do sinal do ruído 

    plt.subplot(3, 1, 3); # definindo nº de linhas, nº de colunas e index do grafico do sinal de saída 
    plt.plot(x, sinal_recebido, 'blue'); # definindo valores do eixo "x" = x e "y" = sinal_recebido_0dB e a cor da linha do grafico do sinal de saída 
    plt.xlabel("Tempo"); # definindo descrição para o eixo x
    plt.ylabel("Amplitude"); # definindo descrição para o eixo y
    plt.title("Sinal de saída par SNR=" + str(SNR) + "dB"); # definindo título para o gráfico do sinal de saída 

    plt.tight_layout(); # ajustando layout dos gráficos para caber todos na imagem

    return sinal, sinal_recebido, ruido;
```
