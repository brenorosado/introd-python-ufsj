## ***Introdução aos Diodos***

Os diodos são componentes eletrônicos fundamentais na eletrônica. Eles possuem <br>propriedades únicas que os tornam extremamente úteis em várias aplicações.

### ***O que é um Diodo?***

Um diodo é um componente eletrônico que permite que a corrente elétrica flua<br> em apenas uma direção. Ele possui duas terminais: o ânodo (positivo) e o cátodo (negativo).

### ***Característica do Diodo***

#### ***Equação de Shockley***
A corrente através de um diodo é descrita pela equação de Shockley:

$I = I_s \cdot (e^{(40 \cdot V)} - 1)$

onde:

$I$ é a corrente através do diodo.<br>
$I_s$ é a corrente de saturação reversa.<br>
$V$ é a tensão através do diodo.<br>

```python
import matplotlib.pyplot as plt
import numpy as np

# Parâmetros do diodo
Is = 1e-6  # Corrente de saturação reversa (A)
n = 1.5    # Fator de idealidade
Vt = 0.025 # Tensão térmica (V)

# Criar uma lista de tensão
tensao = np.linspace(0, 1, 100)

# Calcular a corrente através do diodo usando a equação de Shockley
corrente = Is * (np.exp((tensao / (n * Vt))) - 1)

# Configurar o gráfico
plt.figure(figsize=(10, 5))
plt.plot(tensao, corrente, label='Curva do Diodo')
plt.xlabel('Tensão (V)')
plt.ylabel('Corrente (A)')
plt.title('Característica do Diodo')
plt.grid(True)
plt.legend()
plt.show()
```
