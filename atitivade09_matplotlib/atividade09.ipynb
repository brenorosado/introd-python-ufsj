### Exerc√≠cio 01

```python
import numpy as np; # import biblioteca Numpy como 'np'
import matplotlib.pyplot as plt; # importando biblioteca matplotlib.pyplot como 'plt'
x = np.linspace(0, 10, 10); # 10 pontos do intervalo [0,10)
z = np.linspace(0, 10, 500); # 500 pontos do intervalo [0,10)
y = np.sin(x); # gerando valores do seno para o intervalo e steps defindos em x
w = np.sin(z); # gerando valores do seno para o intervalo e steps defindos em z
plt.subplot(2, 2, 1); # definindo n¬∫ de linhas, n¬∫ de colunas e index do grafico 1
plt.plot(x, y, 'red'); # definindo valores do eixo "x" = x e "y" = y e a cor da linha do grafico 1
plt.title("Seno (10 pontos)"); # definindo o titulo do gr√°fico 1
plt.subplot(2, 2, 2); # definindo n¬∫ de linhas, n¬∫ de colunas e index do grafico 2
plt.plot(z, w, 'blue'); # definindo valores do eixo "x" = z e "y" = w e a cor da linha do grafico 2
plt.title("Seno (500 pontos)"); # definindo o titulo do gr√°fico 2
# Conclus√£o: a linha com mais pontos fica mais fiel ao verdadeiro formato da onda
```

### Exerc√≠cio 02

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 500) # 500 pontos do intervalo [0,10) 
y = np.tan(x) # criando os valores de tangente para os 500 pontos
z = np.cos(x) # criando os valores de coseno para os 500 pontos
ax.plot(x, y, 'b', linewidth = 3, label=r'$y=\tan(x)$'); # plotando o grafico da tangente em azul
ax.plot(x, z, 'r', linewidth = 3, label=r'$z=\cos(x)$'); # plotando o grafico do cosseno em vermelho
ax.legend(loc = 'upper right'); # definindo o lugar da legenda (canto superior direito) 
ax.set_ylim([-4, 4]); # definindo limites para o eixo y para que a funcao tangente nao interfira na escala do grafico
ax.set_title("Fun√ß√µes coseno e tangente") # definindo um titlo para o gr√°fico

fig.savefig('tan_cos.png') # salvando o gr√°fico como uma imagem PNG
```

### Exerc√≠cio 3

```python
x = np.linspace(0, 5, 25); # 25 pontos (step de 0.2) do intervalo [0,5)
f = np.power(x, 1); # gerando valores de uma reta para o array de numeros criado em x
g = np.power(x, 2); # gerando valores quadrados para o array de numeros criado em x
h = np.power(x, 3); # gerando valores ao cubo para o array de numeros criado em x
plt.subplot(2, 2, 1); # definindo n¬∫ de linhas, n¬∫ de colunas e index do grafico da reta
plt.plot(x, f, 'red', linestyle='--', ); # definindo valores do eixo "x" = x e "y" = f e a cor da linha do grafico da reta
plt.title(r'$f(x)=x$'); # definindo o titulo do gr√°fico da reta
plt.subplot(2, 2, 2); # definindo n¬∫ de linhas, n¬∫ de colunas e index do grafico da funcao quadrada
plt.plot(x, g, 'blue', marker='s'); # definindo valores do eixo "x" = x e "y" = g e a cor da linha do grafico da funcao quadrada
plt.title(r'$g(x)=x^2$'); # definindo o titulo do gr√°fico da funcao quadrada
plt.subplot(2, 2, 3); # definindo n¬∫ de linhas, n¬∫ de colunas e index do grafico da funcao cubica
plt.plot(x, h, 'green', marker="v"); # definindo valores do eixo "x" = z e "y" = h e a cor da linha do grafico da funcao cubica
plt.title(r'$h(x)=x^3$'); # definindo o titulo do gr√°fico da funcao cubica
plt.tight_layout(); # ajustando layout dos gr√°ficos para caber todos na imagem
```
